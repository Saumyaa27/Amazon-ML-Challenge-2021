```python
# Importing the required libraries
import numpy as np
import re 
import nltk

nltk.download('stopwords')
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
stop=stopwords.words(fileids='english')
le = WordNetLemmatizer()
```

```python
def process_txt(txt):
    if pd.isna(txt):
      return txt
    new_txt = txt.lower()
    new_txt = re.sub(r'<[\/\w]*>', ' ', new_txt)
    new_txt = re.sub(r'inch', '',new_txt) 
    new_txt = re.sub(r'[^a-z\s\-]', ' ', new_txt) 
    new_txt = new_txt.split(" ")
    new_txt = list(map(lambda x: le.lemmatize(x), new_txt))
    new_txt = list(map(lambda x: x.strip(), new_txt))
    if '' in new_txt:
        new_txt.remove('')
    new_txt = [word for word in new_txt if word not in stop]
    new_txt = [word for word in new_txt if len(word) > 2]
    new_txt = " ".join([word for word in new_txt])
    return new_txt
```

```python
train['TITLE'] = train['TITLE'].apply(process_txt)
test['TITLE'] = test['TITLE'].apply(process_txt)
```

```python
train['DESCRIPTION'] = train['DESCRIPTION'].apply(process_txt) 
test['DESCRIPTION'] = test['DESCRIPTION'].apply(process_txt) 
```

```python
train['BULLET_POINTS'] = train['BULLET_POINTS'].apply(process_txt)
test['BULLET_POINTS'] = test['BULLET_POINTS'].apply(process_txt)
```

```python
import re
i = 0
def common_words_train(x):
  global i
  s1 = x[0]
  s2 = x[1]
  s3 = x[2]

  set1 = set(s1.split())
  set2 = set(s2.split())
  set3 = set(s3.split())

  s = (set1 & set2).union((set2 & set3).union(set1 & set3))
  i += 1
  print(i)
  if not pd.isna(x[3]):
    s.add(x[3].lower())
  t = list(s) 
  
  if len(t) == 0:
    return s1+" " + s2 + " " + s3

  return " ".join([word for word in t if len(word) > 1]) 
```

```python
import re
i = 0
def common_words_test(x):
  global i
  s1 = x[1]
  s2 = x[2]
  s3 = x[3]

  set1 = set(s1.split())
  set2 = set(s2.split())
  set3 = set(s3.split())

  s = (set1 & set2).union((set2 & set3).union(set1 & set3))
  i += 1
  print(i)
  if not pd.isna(x[3]):
    s.add(x[3].lower())
  t = list(s) 
  
  if len(t) == 0:
    return s1+" " + s2 + " " + s3

  return " ".join([word for word in t if len(word) > 1]) 
```

```python
train["Merged"] = train.apply(common_words_train,axis=1)
test["Merged"] = test.apply(common_words_test,axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer(max_features=10000,binary=True)
x1 = v.fit_transform(x)
x1
```
